We can pull data directly from quandl using the `quandl.get_table()`

```python
# Pull the indicators file 
# (indicators are the column names for all the datasets)
df = quandl.get_table('SHARADAR/INDICATORS')
df[df.table=='SF2']
```

```python
# Pull insider trades for Apple
df = quandl.get_table('SHARADAR/SF2', ticker='AAPL')
df.sort_values('filingdate')
```

```python
# Pull prices for Apple
df = quandl.get_table('SHARADAR/SEP', ticker='AAPL')
df.sort_values('date')
```

```python
# Multiple tickers
df = quandl.get_table('SHARADAR/SF2', ticker=['IBM','AAPL'], paginate=True)
df
```

### Subsetting on dates

- gte: greater than or equal
- gt:  greater than or equal
- lte: less than or equal
- lt:  less than

```python
# As of a date
df = quandl.get_table('SHARADAR/SF1', calendardate='2015-12-31', ticker='AAPL')
df
```

```python
# Since a date
df = quandl.get_table('SHARADAR/SF1', calendardate={'gte': '2015-12-31'}, ticker='AAPL')
df
```

```python
# Between dates
df = quandl.get_table('SHARADAR/SF1', calendardate={'gte': '2015-12-31', 'lte': '2017-12-31'}, ticker='AAPL')
df
```

### Subsetting columns

```python
col_list = ['ticker', 'filingdate', 'transactiondate', 'formtype', 'issuername', 'ownername','officertitle', 'securityadcode', 'transactionshares', 'transactionpricepershare']
df = quandl.get_table('SHARADAR/SF2', 
    qopts={'columns':col_list},
    filingdate={'gte': '2015-12-31', 'lte': '2017-12-31'}, ticker='AAPL')
df
```

### Subsetting on dimension

AR = As reported (ARY=annual; ARQ=Quarterly, ART=Trailing 12 months)

MR = most-recent-reported (MRY=annual; MRQ=Quarterly, MRT=Trailing 12 months)

```python
df = quandl.get_table('SHARADAR/SF1', dimension='ARY', ticker='AAPL')
df
```

### Corporate actions

```python
df = quandl.get_table('SHARADAR/ACTIONS', ticker=['IBM','AAPL','TSLA'])
df.action.value_counts()
```

```python
# List of actions
df=quandl.get_table('SHARADAR/INDICATORS', table='ACTIONTYPES')
df
```

### Corporate events

```python
# events by ticker
df = quandl.get_table('SHARADAR/EVENTS', ticker=['IBM','AAPL','TSLA'])
df
```

```python
# events by date
df = quandl.get_table('SHARADAR/EVENTS', date={'gte':'2022-01-01'}, ticker=['IBM','AAPL','TSLA'])
df
```

```python
# List of events
df=quandl.get_table('SHARADAR/INDICATORS', table='EVENTCODES')
df
```

```python
# 13D events
df = quandl.get_table('SHARADAR/EVENTS', ticker=['IBM','GME','TSLA','BBBY'])
df['activist']=(df.eventcodes.str.find('35')>=0)
df[df.activist==True]
```

```python
# Downloading zipped tables
for table in ['INDICATORS','TICKERS','SF1','DAILY','SEP','SF2','SF3','SFP'] :
    quandl.export_table('SHARADAR/'+table)
    print("finished", table)
```

```python
# Unzipping and loading full dataset
from zipfile import ZipFile
table = "SF2"
with ZipFile('./SHARADAR_'+table+'.zip', 'r') as zipObj:
    name = zipObj.namelist()[0]
    zipObj.extractall()
df = pd.read_csv(name, low_memory=False)
df = df.rename(columns={"table": "table_"})
```
